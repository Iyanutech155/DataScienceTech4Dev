```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

```

```python
data=pd.read_csv('../input/fifa-world-cup/WorldCups.csv')
data
```

```python
plt.figure(figsize=(14,8))
plt.ylabel('Number')
plt.xlabel('Countries')
plt.title('Fifa frequency of winners from 1982 to 2014')

sns.barplot(x= data.Country, y=data.GoalsScored)


```
